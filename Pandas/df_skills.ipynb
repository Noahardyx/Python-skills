```python
import pandas as pd
import numpy as np
```

```python
# 0. Create df
df = pd.DataFrame({'country':['Australia','Japan','China','Korea','United States','France','United Kingdom','India'],
                   'gdp':[111,222,333,444,555,666,777,888],
                   'inflation':['11$','22$','33$','44$','55$','66$','77$','88$']})

df1 = pd.DataFrame({'country':['Australia','Japan','China','Korea','United States','France','United Kingdom','India'],
                   'gdp':[111,222,333,444,555,666,777,888],
                   'inflation':['188,835,201,625.910 (US$)','tp','188,835,201,625.910 (US$)','188,835,201,625.910 (US$)','188,835,201,625.910 (US$)','188,835,201,625.910 (US$)','188,835,201,625.910 (US$)','188,835,201,625.910 (US$)']})

df1 = df1.replace('tp', np.NaN)

df
```

```python
# 1. Extract only one element

df[df['country']=='Korea']
```

```python
# 2. Extract more two element with list

asia = ['Japan', 'China', 'Korea', 'India']

asia_df = pd.DataFrame()

for i in asia:
  asia_df = asia_df.append(df[df['country']==i], ignore_index=True)

asia_df
```

```python
# 3. Column Name Change

df.columns = ['국가', 'GDP', 'Inflation']
df
```

```python
# 4. Df Sort

df = df.sort_values('국가')
df = df.reset_index(drop=True)

df
```

```python
df
```

```python
# 5. String -> Int

df['Inflation'] = pd.to_numeric(df['Inflation'].str.replace("$", ""))

df
```

```python
# NaN값 제거

df1 = df1.dropna(axis=0)
df1 = df1.reset_index(drop=True)

df1
```

```python
# 5-1 String -> Int

# df1['inflation'] = df1['inflation'].str.replace(" (US$)", "")

for i in range(len(df1)):
  df1['inflation'][i] = df1['inflation'][i][:-10]

df1["inflation"] = pd.to_numeric(df1["inflation"].str.replace(',', ''))

df1
```
